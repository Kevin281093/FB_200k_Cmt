```python
# Read data from csv
import csv
cmt = []
with open (r"C:\Users\ASUS\Downloads\FB_200k_Cmt\200k_comments.csv", mode= "r", encoding= 'utf8') as file:
    for i in file: # Filter data with conditions then add to array name data
        if ("công vinh" in i.lower() and "từ thiện" in i.lower()) or ("thuỷ tiên" in i.lower() and "từ thiện" in i.lower()):
            cmt.append(i.strip())
```

```python
# Load data to database (SQL Server)
try:
    import pyodbc as odbc
    driver = "ODBC Driver 17 for SQL Server"
    server = "LAPTOP-VE5ATOI8\SQLEXPRESS"
    database = "FB_200k_Cmt"
    conn = odbc.connect(f"DRIVER={driver}; SERVER={server};DATABASE={database}; Trusted_Connection=yes;")
    cur = conn.cursor()
    #Drop table if it exists
    drop_script = """IF OBJECT_ID('dbo.FB_200k_Cmt', 'U') IS NOT NULL 
                    DROP TABLE dbo.FB_200k_Cmt """
    cur.execute(drop_script)
    #Create table if it not exists
    create_script = """IF OBJECT_ID('dbo.FB_200k_Cmt', 'U') IS NULL
        CREATE TABLE dbo.FB_200k_Cmt (comment nvarchar(max))""" #Create collumn with nvarchar with max length and define UTF8
    cur.execute(create_script)
    for i in cmt:
        print(i)
        insert_script = f"""INSERT INTO dbo.FB_200k_Cmt (comment) 
                        VALUES (?)"""
        cur.execute(insert_script,i)
    conn.commit()
except Exception as err:
    print(err)
finally:
    if cur is not None:
        cur.close()
    if conn is not None:
        conn.close()
```
